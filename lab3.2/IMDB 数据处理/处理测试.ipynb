```python
import pandas as pd

data_m = pd.read_csv(
    filepath_or_buffer='/Users/fox/Library/CloudStorage/OneDrive-PetertheGreatSt.PetersburgPolytechnicalUniversity/СПБПУ/3 курс/6 семестр/СУБД/资料/DataSet/data_actresses.list.txt',
    header=0,
    sep='\t'
)
data_m.columns = ['name', 't1', 't2', 't3']
```

```python
data_m['t1'].fillna(value='', inplace=True)
data_m['t2'].fillna(value='', inplace=True)
data_m['t3'].fillna(value='', inplace=True)

moive_list = data_m['t1'] + data_m['t2'] + data_m['t3']

data_m = pd.concat([data_m['name'], moive_list], axis=1)
data_m.columns = ['name', 'title_mix']
```

```python
data_m
```

```python
import numpy as np
import re
```

```python
data_m = data_m.iloc[:1000, :]
```

```python
name_title = []
for col in data_m.values:
    if (col[0] is not np.nan):
        names = col[0]
        name_list = names.split(', ')
        

    for this_name in name_list:

        title_mix = col[1]
        title = None
        year = None
        series_name = None  # 系列名称： {}
        character_name = None  # 角色名称

        title = re.search(r'^[^\(\{\[]*', title_mix)
        if title is not None:
            title = str(title.group()[:-1])

        year = re.search(r'(?!=\({1})[\d]{4}(?!=\){1})', title_mix)
        if year is not None:
            year = int(year.group())

        series_name =  re.search(r'\{(.*?)\}', title_mix)
        if series_name is not None:
            series_name = str(series_name.group()[1:-1])

        character_name =  re.search(r'\[(.*?)\]', title_mix)
        if character_name is not None:
            character_name = str(character_name.group()[1:-1])
        # name_title.append([this_name, title, series_name, year, character_name])
        # name_title.append([this_name, [title, series_name, year, character_name]])
        rols = pd.DataFrame([[title, series_name, year, character_name]], 
                            columns=['title', 'series name', 'year', 'character name'])

        name_title.append([this_name, rols])
        
```

```python
name_title = pd.DataFrame(name_title, columns=['name', 'rols'])
name_title.sort_values(by='name', inplace=True)
name_title.reset_index(drop=True, inplace=True)
name_title.head(20)
```
